```python
import pandas as pd 
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
from sklearn.svm import LinearSVC
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.model_selection import train_test_split
```

```python
data=pd.read_csv("trainAndTest.csv")
```

```python
data
```

```python
cevir=LabelEncoder()
data["diagnosis"]=cevir.fit_transform(data["diagnosis"])
data
```

```python
train,test=train_test_split(data,test_size=0.2)
```

```python
train
```

```python
xTrain=train.drop("diagnosis",axis=1)
yTrain=train.loc[:,"diagnosis"]
```

```python
xTest=test.drop("diagnosis",axis=1)
yTest=test.loc[:,"diagnosis"]
```

```python
model=LogisticRegression()
model.fit(xTrain,yTrain)
```

```python
tahminler=model.predict(xTest)
```

```python
confusion_matrix(yTest,tahminler)
```

```python
print(classification_report(yTest,tahminler))
```

```python
model2=LinearSVC()
model2.fit(xTrain,yTrain)
```

```python
tahminler2=model2.predict(xTest)
```

```python
confusion_matrix(yTest,tahminler2)
```

```python
print(classification_report(yTest,tahminler2))
```
