```python
import pandas as pd 
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.svm import LinearSVC
```

```python
data = pd.read_csv("trainAndTest.csv")
```

```python
sayiyaCevir=LabelEncoder()
data["diagnosis"]=sayiyaCevir.fit_transform(data["diagnosis"].values)
data
```

```python
train,test=train_test_split(data,test_size=0.2)
```

```python
train
```

```python
test
```

```python
xTrain=train.drop("diagnosis",axis=1)
yTrain=train.loc[:,"diagnosis"]
xTrain
```

```python
xTest=test.drop("diagnosis",axis=1)
yTtest=test.loc[:,"diagnosis"]
xTest
```

```python
model=LogisticRegression()
model.fit(xTrain,yTrain)
```

```python
tahminler=model.predict(xTest)
tahminler
```

```python
confusion_matrix(yTtest,tahminler)
```

```python
print(classification_report(yTtest,tahminler))
```

```python
model2=LinearSVC()
model2.fit(xTrain,yTrain)

```

```python
tahminler=model2.predict(xTest)
```

```python
tahminler
```

```python
confusion_matrix(yTtest,tahminler)
```

```python
print(classification_report(yTtest,tahminler))
```
