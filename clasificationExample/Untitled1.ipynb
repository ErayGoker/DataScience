```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix
from sklearn.svm import LinearSVC
```

```python
data=pd.read_csv("trainAndTest.csv")
data
```

```python
encoder=LabelEncoder()
data["diagnosis"]=encoder.fit_transform(data["diagnosis"])
data
```

```python
train,test=train_test_split(data,test_size=0.2)
test
```

```python
xTrain=train.drop("diagnosis",axis=1)
yTrain=train.loc[:,"diagnosis"]
```

```python
xTest=test.drop("diagnosis",axis=1)
yTest=test.loc[:,"diagnosis"]
```

```python
model=LogisticRegression()
model.fit(xTrain,yTrain)
```

```python
tahminler=model.predict(xTest)
tahminler
```

```python
confusion_matrix(yTest,tahminler)
```

```python
print(classification_report(yTest,tahminler))
```

```python
model2=LinearSVC()
model2.fit(xTrain,yTrain)
```

```python
tahminler2=model2.predict(xTest)
```

```python
tahminler2
```

```python
confusion_matrix(yTest,tahminler2)
```

```python
print(classification_report(yTest,tahminler2))
```
