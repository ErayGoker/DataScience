```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
train=pd.read_csv("train.csv")
test=pd.read_csv("test.csv")
test
```

```python
xTrain=train.drop("SalePrice",axis=1)
yTrain=train.loc[:,"SalePrice"]
yTrain
```

```python
model=LinearRegression()
model.fit(xTrain,yTrain)
```

```python
xTest=test.drop("SalePrice",axis=1)
yTest=test.loc[:,"SalePrice"]
yTest
```

```python
tahminler=model.predict(xTest)
dogrulukOrani=np.sqrt(mean_squared_error(yTest,tahminler))
dogrulukOrani
```
