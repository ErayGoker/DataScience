```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv("DecisionTrees_titanic.csv")
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.2)
```

```python
trainx=train.drop("Survived",axis=1)
trainy=train.loc[:,"Survived"]
testx=test.drop("Survived",axis=1)
testy=test.loc[:,"Survived"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

model1=LogisticRegression()
model2=LinearSVC()
model3=DecisionTreeClassifier()
model4=RandomForestClassifier()

model1.fit(trainx,trainy)
model2.fit(trainx,trainy)
model3.fit(trainx,trainy)
model4.fit(trainx,trainy)
```

```python
tahminler1=model1.predict(testx)
tahminler2=model2.predict(testx)
tahminler3=model3.predict(testx)
tahminler4=model4.predict(testx)
```

```python
from sklearn.metrics import classification_report 
print(classification_report(testy,tahminler1))
```

```python
print(classification_report(testy,tahminler2))
```

```python
print(classification_report(testy,tahminler3))
```

```python
print(classification_report(testy,tahminler4))
```

```python
import matplotlib.pyplot as plt
from sklearn.tree import plot_tree
```

```python
x=pd.concat([trainx,testx],axis=0)
listx=list(x)
plt.figure(figsize=(12,12))
plot_tree(model3,
        max_depth=5,
        feature_names=listx,
        filled=True,
        impurity=False,
        rounded=True,
        precision=1
         )
plt.show()
```
