```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression,Ridge
from sklearn.metrics import mean_squared_error
```

```python
train=pd.read_csv("train.csv")
```

```python
train
```

```python
train,test=train_test_split(train,test_size=0.2)
```

```python
test
```

```python
train
```

```python
xTrain=train.drop("SalePrice",axis=1)
yTrain=train.loc[:,"SalePrice"]
```

```python
xTest=test.drop("SalePrice",axis=1)
yTest=test.loc[:,"SalePrice"]
```

```python
model1=LinearRegression()
model1.fit(xTrain,yTrain)
tahminler=model1.predict(xTest)
mean_squared_error(yTest,tahminler)
```

```python
model2=Ridge(alpha=5)
model2.fit(xTrain,yTrain)
tahminler2=model2.predict(xTest)
mean_squared_error(yTest,tahminler2)
```
