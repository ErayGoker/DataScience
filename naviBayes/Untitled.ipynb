```python
import pandas as pd 
import numpy as np
```

```python
data=pd.read_csv("breast-cancer.csv")
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["diagnosis"]=le.fit_transform(data["diagnosis"])
```

```python
data
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(data,test_size=0.2)
```

```python
trainX=train.drop("diagnosis",axis=1)
trainY=train.loc[:,"diagnosis"]
testX=test.drop("diagnosis",axis=1)
testY=test.loc[:,"diagnosis"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import GaussianNB

model1=LogisticRegression()
model2=LinearSVC()
model3=GaussianNB()

model1.fit(trainX,trainY)
model2.fit(trainX,trainY)
model3.fit(trainX,trainY)

tahminler1=model1.predict(testX)
tahminler2=model2.predict(testX)
tahminler3=model3.predict(testX)


```

```python
from sklearn.metrics import classification_report,confusion_matrix

confusion_matrix(testY,tahminler1)
```

```python
confusion_matrix(testY,tahminler2)
```

```python
confusion_matrix(testY,tahminler3)
```

```python
print(classification_report(testY,tahminler1))
```

```python
print(classification_report(testY,tahminler2))
```

```python
print(classification_report(testY,tahminler3))
```
